```python
#Filter the data 
female_survived_age_above_31=df[(df["Sex"]=="female") & (df["Survived"]==1) & (df["Age"]>18) & (df["Age"]<65)]

#Count the number of passengers
count=female_survived_age_above_31

print ("Number of Female Passenger Who survived & were older than 31;",count)
```

```python
#filter the data
male_survived_age_above_31=df[(df["Sex"]=="male") & (df["Survived"]==1) & (df["Age"]>18) & (df["Age"]<65)]

#Count the number of passengers
count=male_survived_age_above_31

print (count)
```

```python
#count the number of female passengers who survived & were older than 18
female_count=female_survived_age_above_31.shape[0]

#count the number of female passengers who survived & were older than 18
male_count=male_survived_age_above_31.shape[0]

#prepare data for plotting 
counts=pd.DataFrame({
    'Sex':['Female','Male'],
    'Survived_Age_Above_31':[female_count,male_count]
})

#Create a bar plot 
plt.figure(figsize=(8,6))
sns.barplot(x='Sex',y='Survived_Age_Above_31',data=counts,palette='viridis')
plt.title('Number of Passengers Who Survived & Were Older than 18')
plt.xlabel('Sex')
plt.ylabel('Count')
plt.Show()
```
