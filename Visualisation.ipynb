```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
df=pd.read_csv(r"car.csv")
```

```python
#df.head(20)
df.tail(20)
```

# for unwanted Data

```python
df["normalized-losses"]=df['normalized-losses'].replace('?',np.nan)
df
```

```python
df["normalized-losses"]=df['normalized-losses'].astype(float)
df
```

```python
df["normalized-losses"]=df['normalized-losses'].fillna(df["normalized-losses"].mean())
df
```

```python
#FOR UNWANTED DATA
df['horsepower']=df["horsepower"].replace('?',np.nan)
df
df["horsepower"]=df['horsepower'].astype(float)
df
df["horsepower"]=df['horsepower'].fillna(df["horsepower"].mean())
df
```

```python
plt.figure(figsize=(8,6))
plt.hist(df['width'],bins=20)
plt.title("Width frequency distribution")
plt.xlabel("frequency")
plt.ylabel("width")

plt.show()
```

```python
# In statistics, data is usually sorted in one way or another. 
# You might sort the data into classes, categories, 
# by range or placement on the number line.
# A bin—sometimes called a class interval—is a way of sorting data in a 
# histogram.
# It’s very similar to the idea of putting data into categories.

plt.figure(figsize=(8,6))
plt.hist(df['width'],bins=20)
# df['width'].hist(bins=10)
plt.title("Width frequency distribution")
plt.xlabel("frequency")
plt.ylabel("width")
plt.show()
```

```python
#plot Mean with Median
width_mean=df['width'].mean()
width_median=df['width'].median()
plt.figure(figsize=(8,6))
#plt.hist(df['width'],bins=20)
df['width'].hist(bins=10)
plt.title("Width frequency distribution")
plt.axvline(width_mean,color='red')  #axis of verticle line
plt.axvline(width_median,color='yellow')
plt.xlabel("frequency")
plt.ylabel("width")
plt.show()
```
